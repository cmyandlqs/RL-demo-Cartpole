```python
import random
import collections

# 定义一个简单的状态类，用于演示
class State:
    def __init__(self, value):
        self.value = value
    
    def __repr__(self):
        return f"State({self.value})"

# 创建示例缓冲区
buffer = collections.deque(maxlen=100)  # 使用你提到的 deque

# 添加一些示例经验转换 (state, action, reward, next_state)
for i in range(10):
    state = State(f"state_{i}")
    action = random.randint(0, 1)  # 动作通常是 0 或 1
    reward = random.uniform(-1, 1)  # 奖励通常是浮点数
    next_state = State(f"next_state_{i}")
    
    transition = (state, action, reward, next_state)
    buffer.append(transition)

print("完整缓冲区:")
for i, trans in enumerate(buffer):
    print(f"{i}: {trans}")

print("\n" + "="*50)

# 随机采样 3 个经验转换
batch_size = 3
transitions = random.sample(list(buffer), batch_size)

print(f"\n随机采样的 {batch_size} 个经验转换:")
for i, trans in enumerate(transitions):
    print(f"Sample {i+1}: {trans}")

# 验证采样的特性：无重复且不改变原缓冲区
print(f"\n原缓冲区长度: {len(buffer)}")
print(f"采样结果长度: {len(transitions)}")
print(f"采样结果是否包含重复项: {len(transitions) != len(set(transitions))}")
```

```python
'''
*transitions 称为解包操作符，它将 transitions 列表中的每个元素作为单独的参数传递给 zip() 函数；
zip() 函数将多个可迭代对象打包成一个元组序列
'''
# 示例 transitions 数据
transitions = [
    ('state1', 'action1', 10, 'next1', True),
    ('state2', 'action2', -5, 'next2', False),
    ('state3', 'action3', 20, 'next3', True)
]

# 执行这行代码
state, action, reward, next_state, done = zip(*transitions)
print(f'transitions: {transitions}')
print("state:", state)         # ('state1', 'state2', 'state3')
print("action:", action)       # ('action1', 'action2', 'action3')
print("reward:", reward)       # (10, -5, 20)
print("next_state:", next_state)  # ('next1', 'next2', 'next3')
print("done:", done)           # (True, False, True)
```
